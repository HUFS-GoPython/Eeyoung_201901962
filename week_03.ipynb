## 논리연산자
- 비교연산자가 여러번 사용될 때 사용하는 것
- and, or, not
    - A and B: A도 참이고, B도 참일 때 True
        - 예: A & B
    - A or B: A 혹은 B, 둘중 하나만 참이면 True
        - 예: A | B
    - not A: A가 False -> True, True -> False
- 비교연산자의 괄호는 가독성을 위해 (optional)
- 파이썬에서만 가능한 표현: 0 < x < 10

### True, False
- "만약 a가 참이라면, ..."
- a = True

```python
if a == True:
    print() #이렇게 쓰면 틀림
if a:
    print() #이렇게 써야됨
```

## in
- membership operator
- 어떤 요소가 그 안에 있는지 확인하고 싶다
- 비슷한 조건이 여러번 반복될 때 사용가능

```python
letter = 'k'

# letter가 모음인지 확인하고 싶을 때
#if letter == 'a' or letter == 'e'...
if letter in ['a', 'e', 'i', 'o', 'u']: # 리스트 요소에 letter와 일치하는 것이 있는가
    print('Yes')
else:
    print('No')
```

```python
word = 'abcde'

# word에 모음이 있는지 확인
if 'a' in word:
    print('Yes')
elif 'e' in word:
    print('Yes')
elif 'i' in word:
    print('Yes')
elif 'o' in word:
    print('Yes')
elif 'u' in word:
    print('Yes')
else:
    print('No')
```

## 바다코끼리 연산자
- walrus operator
- 할당과 테스트를 동시에 하는 기능
- :=

```python
tweet_limit = 200
tweet_string = 'blah' * 50
tweet_string
```

```python
diff = tweet_limit - len(tweet_string)
if diff > 0:
    print("가능")
else:
    print("불가능")
```

```python
if diff := tweet_limit - len(tweet_string) > 0:
    print("가능")
```

## Quiz
- 윤년을 구하는 코드를 짜보자
    - 연도가 4로 나눠떨어지면 윤년
    - 100으로 나눠떨어지면 안된다
    - 400으로 나눠떨어지면 된다
- year를 입력받는다
- 출력= '{}년은 윤년입니다' / "{}년은 윤년이 아닙니다."

```python
year = int(input())
if ((year % 4) == 0 and (year % 100) != 0) or year % 400 == 0:
    print(f'{year}년은 윤년입니다')
else:
    print(f'{year}년은 윤년이 아닙니다')
```

## Slicing
- 지정 범위의 문자열을 추출하는 것
- 문법: 대괄호 + 오프셋 이용 [offset]
    - [start:end:step]
    - start 포함, end 불포함 (크거나 같고, 작다)
    - 숫자 하나만 쓰면 start
- 오프셋은 항상 0부터 시작
- 0일 경우 생략한다
    - 예: ~4까지 출력하라 => [:5]
    - 처음부터 끝까지 2스텝씩 => [::2]
- 오프셋은 음수로도 가능하다 (-1부터 시작)

## - set
    - {a, b, c, ...}
    - 중복 불가능
    - 수학의 집합과 동일한 개념이다
    - 순서 없음 : 슬라이싱 못함, 몇번째 없음
- list
    - [a, b, c, ...]
    - 중복 가능
    - 순서 있음

```python
cars = ['kia', 'benz', 'porsche', 'kia']
set(cars)
```

```python
# lexical diversity
len(set(moby.tokens)) / len(moby.tokens)
```

```python
# 오름차순, 내림차순 정렬
sorted(cars, reverse = True)
#문장부호 -> 숫자 -> 대문자 -> 소문자
```

```python
sorted(moby.tokens, reverse=True)[:20]
```
